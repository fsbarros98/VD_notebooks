```r
library(ggplot2)
library(maps)
library(mapproj)
```

```r
states_map <- map_data("state")
```

```r
states_map
```

```r
ggplot(states_map, aes(x=long, y=lat, group=group)) +
    geom_polygon(fill="white", color="black")
```

```r
ggplot(states_map, aes(x=long, y=lat, group=group)) +
    geom_path() +
    coord_map("mercator")
```
