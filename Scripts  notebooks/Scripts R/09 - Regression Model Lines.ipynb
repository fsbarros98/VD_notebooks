```r
# To create a line representing the regression line to a scatter we add stat_smooth() with the method=lm
```

```r
sp <- ggplot(heightweight, aes(x=ageYear, y=heightIn))
```

```r
sp + geom_point() + stat_smooth(method=lm)
```

```r
# By default, stat_smooth() also adds a 95% confidence region for the regression fit
# This interval can be changed using level or disbled with se=FALSE
```

```r
# Let's increase the confidence level
sp + geom_point() + stat_smooth(method=lm, level=0.999)
```

```r
# And, with no confidence region and the regresison line in red
sp + geom_point() + stat_smooth(method=lm, se=FALSE, color="red")
```

```r
# Note that the linear regression is not the only way of fitting a model to the data
# In fact, it is not even the default one
sp + geom_point() + stat_smooth()  # Obs: stat_smooth() is the same as stat_smooth(method=loess)
```
